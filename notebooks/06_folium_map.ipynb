```python
url = "https://raw.githubusercontent.com/pakizaim7/spacex-capstone/main/notebooks/spacex_launch_data_cleaned.csv"
df = pd.read_csv(url)
df.head()
```

```python
launch_sites_coords = {
    'CCAFS LC-40': [28.5623, -80.5774],
    'VAFB SLC-4E': [34.6328, -120.6108],
    'KSC LC-39A': [28.5733, -80.6469],
    'CCAFS SLC-40': [28.5623, -80.5774]
}
```

```python
# Centering map at Florida
spacex_map = folium.Map(location=[28.5623, -80.5774], zoom_start=5)

# Adding launch site markers
for site, coord in launch_sites_coords.items():
    folium.Marker(
        location=coord,
        popup=site,
        icon=folium.Icon(color='red', icon='rocket', prefix='fa')
    ).add_to(spacex_map)

# Displaying the map
spacex_map
```

```python
for index, row in df.iterrows():
    coord = launch_sites_coords.get(row['launchpad'], None)
    if coord:
        color = 'green' if row['success'] == 1 else 'red'
        folium.CircleMarker(
            location=coord,
            radius=6,
            color=color,
            fill=True,
            fill_opacity=0.7,
            popup=f"Mission: {row['name']}, Payload: {row.get('payload_mass', 'N/A')} kg"
        ).add_to(spacex_map)
```

```python
spacex_map.save("slide_34_map.html")
```

```python
import folium

# Filtering successful landings (class=1)
success_df = df[df['class'] == 1]

# Centering map roughly at the mean coordinates of launchpads (simulated lat/lon)
spacex_map = folium.Map(location=[28, -80], zoom_start=4)

# Adding markers for each launch site
for _, row in success_df.iterrows():
    folium.Marker(
        location=[28, -80],  # Replace with actual lat/lon if available
        popup=f"{row['launchpad']}\n{row['name']}",
        icon=folium.Icon(color='green')
    ).add_to(spacex_map)

# Saving as HTML
spacex_map.save("slide_35_map.html")
```

```python
# Filtering failed landings (class=0)
fail_df = df[df['class'] == 0]

spacex_map_fail = folium.Map(location=[28, -80], zoom_start=4)

for _, row in fail_df.iterrows():
    folium.Marker(
        location=[28, -80],  # Replace with actual lat/lon if available
        popup=f"{row['launchpad']}\n{row['name']}",
        icon=folium.Icon(color='red')
    ).add_to(spacex_map_fail)

spacex_map_fail.save("slide_36_map.html")
```

```python
spacex_map_all = folium.Map(location=[28, -80], zoom_start=4)

for _, row in df.iterrows():
    color = 'green' if row['class'] == 1 else 'red'
    folium.Marker(
        location=[28, -80],  # Replace with actual lat/lon if available
        popup=f"{row['launchpad']}\n{row['name']}",
        icon=folium.Icon(color=color)
    ).add_to(spacex_map_all)

spacex_map_all.save("slide_37_map.html")
```
