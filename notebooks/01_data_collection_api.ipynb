```python
import requests
import pandas as pd
```

```python
# SpaceX API endpoint for past launches
url = "https://api.spacexdata.com/v4/launches/past"

# Sending GET request
response = requests.get(url)

# Checking response status
print("Status Code:", response.status_code)

# Converting response to JSON
launches = response.json()

# Showing number of records
len(launches)
```

```python
# Converting JSON to pandas DataFrame
df = pd.json_normalize(launches)

# Displaying first 5 rows
df.head()
```

```python
df = df[
    [
        'flight_number',
        'name',
        'date_utc',
        'success',
        'rocket',
        'payloads',
        'launchpad'
    ]
]

df.head()
```

```python
# Converting date column to datetime
df['date_utc'] = pd.to_datetime(df['date_utc'])

# Extracting launch year
df['launch_year'] = df['date_utc'].dt.year

df.head()
```

```python
# Creating class label (1 = Success, 0 = Failure)
df['class'] = df['success'].apply(lambda x: 1 if x == True else 0)

df[['flight_number', 'success', 'class']].head()
```

```python
# Saving dataset to CSV
df.to_csv("spacex_launch_data_api.csv", index=False)

print("CSV file saved successfully!")
```
