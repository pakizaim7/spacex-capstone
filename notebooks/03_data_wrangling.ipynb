```python
import pandas as pd
```

```python
# Loading dataset from GitHub
url = "https://raw.githubusercontent.com/pakizaim7/spacex-capstone/main/notebooks/spacex_launch_data_api.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
# Converting date_utc to datetime
df['date_utc'] = pd.to_datetime(df['date_utc'])

df.dtypes
```

```python
df.isnull().sum()
```

```python
df['success'] = df['success'].fillna(False)
```

```python
# Creating binary class label
df['class'] = df['success'].apply(lambda x: 1 if x == True else 0)

df[['success', 'class']].head()
```

```python
df['launch_year'] = df['date_utc'].dt.year

df[['date_utc', 'launch_year']].head()
```

```python
# Keeping only relevant columns
df = df[
    [
        'flight_number',
        'name',
        'launch_year',
        'success',
        'class',
        'rocket',
        'payloads',
        'launchpad'
    ]
]

df.head()
```

```python
df.info()
df.shape
```

```python
df.to_csv("spacex_launch_data_cleaned.csv", index=False)

print("Cleaned dataset saved successfully!")
```
